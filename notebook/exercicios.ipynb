## Condicional

```python
# A função verifica_se_pode_dirigir verifica se a idade do usuário é maior
# que 18 e imprime se ele tem permissão para dirigir ou não.
def verica_se_pode_dirigir(idade):
    if idade >= 18:
        print('Tem permissão para dirigir')
    else:
        print('Não tem permissão para dirigir')
```

```python
# João não pode dirigir porque tem menos de 18 anos.
idade_joao = 15
verica_se_pode_dirigir(idade_joao)
```

```python
# Maria pode dirigir porque tem 18 anos.
idade_maria = 18
verica_se_pode_dirigir(idade_maria)
```

```python
# Ana pode dirigir porque tem mais de 18 anos.
idade_ana = 45
verica_se_pode_dirigir(idade_ana)
```

## Convertendo de `str` pra `int`

```python
# A função verifica_se_pode_dirigir_sem_parametro solicita a idade do usuário e
# verifica se o usuário é maior que 18. Ela então imprime se ele tem permissão
# para dirigir ou não.
def verifica_se_pode_dirigir_sem_parametro():
    idade = input('Qual sua idade')

    # A variável idade deve ser convertida para um inteiro.
    idade = int(idade)
    if idade >= 18:
        print(f'{idade} anos: tem permissão para dirigir')
    else:
        print(f'{idade} anos: não tem permissão para dirigir')
```

```python
verifica_se_pode_dirigir_sem_parametro()
```

```python
# Podemos selecionar uma parte da lista usando o fatiamento. Basta usar o nome
# da lista e os índices de início e fim, separados por dois-pontos. O primeiro
# índice indica o início do fatiamento e o segundo o fim, sendo que o índice
# não é incluído no fatiamento.

# Aqui, por exemplo, pegamos os elementos da lista de idades, da posição 0 até  a 1.
idades[0:2]
```

```python
# Se só o primeiro índice for informado, o Python assume que o último índice é
# o último elemento da lista.

# Assim serão coletadas todas as idades com excessão da primeira.
idades[1:]
```

```python
# Se usarmos apenas o segundo índice, o Python assume que o primeiro índice é 0
idades[:2]
```

```python
# Para pegar os itens da lista, de trás para frente, basta usar o sinal de menos
idades[-1]
```

## Laços de repetição com `for`

```python
idades = [18, 22, 15, 50 ]
def verifica_se_pode_dirigir(idade):
    if idade >= 18:
        print(f'{idade} anos de idade, TEM permissão para dirigir')
    else:
        print(f'{idade} anos de idade, NÃO TEM permissão para dirigir')

# com a instrução for, podemos iterar sobre uma lista.
for idade in idades:
    verifica_se_pode_dirigir(idade)
```

```python
# Podemos incluir o for dentro da função
def verifica_se_pode_dirigir_com_for(idades):
    for idade in idades:
        if idade >= 18:
            print(f'{idade} anos de idade, TEM permissão para dirigir')
        else:
            print(f'{idade} anos de idade, NÃO TEM permissão para dirigir')


idades = [18, 22, 15, 50 ]
verifica_se_pode_dirigir_com_for(idades)
```

## Booleano

```python
# Uma expressão de comparação é um valor que retorna um valor booleano.
# No exemplo abaixo, a expressão retorna um valor verdadeiro
idade = 18
idade >= 18
```

```python
# Uma expressão de comparação é um valor que retorna um valor booleano.
# No exemplo abaixo, a expressão retorna um valor false
idade = 15
idade >= 18
```

```python
idades = [20, 14, 40]
permissoes = []

# Vamos armazenar, na lista de permissões, se o usuário tem permissão para
# dirigir ou não.
def verifica_se_pode_dirigir_e_armazena_na_lista(idades, permissoes):
    for idade in idades:
        if idade >= 18:
            permissoes.append(True)
        else:
            permissoes.append(False)

verifica_se_pode_dirigir_e_armazena_na_lista(idades, permissoes)
```

```python
permissoes
```

```python
for permissao in permissoes:
    if permissao == True:
        print('Tem permissão para dirigir')
    else:
        print('Não tem permissão para dirigir')
```
