# Lógica de programação com Python

```python
# O resultado de uma soma de dois inteiros é um inteiro.
1+1
```

```python
# Uma string é uma sequência de caracteres.
"Guilherme"
```

```python
# Uma string pode ser atrubuída a uma variável.
nome = 'Guilherme'
```

```python
# Ao referenciar uma variável, o Python retorna o valor dela.
nome
```

```python
# Um inteiro também pode ser atribuído a uma variável.
idade = 28
```

```python
# A funçãpo print imprime o valor de uma variável.
print('O nome é Guilherme e sua idade é 28 anos')
```

```python
# Podemos inserir variáveis dentro de strings.
print(f'O nome é {nome} e sua idade é {idade} anos')
```

```python
# A função saudação solicita o nome do usuário e imprime uma saudação.
def saudacao():
    nome = input('Qual o seu nome?')
    print(f'olá {nome}')
```

```python
# Ao invocar a função saudacao, o usuário é solicitadoa informar o nome.
# A função, então, imprime uma saudação.
saudacao()
```

```python
# Agora vamos criar uma função de saudação que usa um valor de fora da função
nome = 'Luiza'

def saudacao_com_parametros(nome_da_pessoa):
    print(f'Olá {nome_da_pessoa}')

saudacao_com_parametros(nome)
```

```python
# A função verifica_se_pode_dirigir verifica se a idade do usuário é maior
# que 18 e imprime se ele tem permissão para dirigir ou não.
def verica_se_pode_dirigir(idade):
    if idade >= 18:
        print('Tem permissão para dirigir')
    else:
        print('Não tem permissão para dirigir')
```

```python
# João não pode dirigir porque tem menos de 18 anos.
idade_joao = 15
verica_se_pode_dirigir(idade_joao)
```

```python
# Maria pode dirigir porque tem 18 anos.
idade_maria = 18
verica_se_pode_dirigir(idade_maria)
```

```python
# Ana pode dirigir porque tem mais de 18 anos.
idade_ana = 45
verica_se_pode_dirigir(idade_ana)
```

```python
# A função verifica_se_pode_dirigir_sem_parametro solicita a idade do usuário e
# verifica se o usuário é maior que 18. Ela então imprime se ele tem permissão
# para dirigir ou não.
def verifica_se_pode_dirigir_sem_parametro():
    idade = input('Qual sua idade')

    # A variável idade deve ser convertida para um inteiro.
    idade = int(idade)
    if idade >= 18:
        print(f'{idade} anos: tem permissão para dirigir')
    else:
        print(f'{idade} anos: não tem permissão para dirigir')
```

```python
verifica_se_pode_dirigir_sem_parametro()
```
